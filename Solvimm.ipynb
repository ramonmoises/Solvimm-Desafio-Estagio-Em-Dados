**1.3. Quais os 5 anos com menos lançamentos de filmes?**

```python
movies = pd.read_csv("movies.csv", sep=",", names=('ID/Filmes', "Ano")) #lendo o arquivo base, atribuiundo um nome ao arquivo, separando o ANO e atribuindo nomes as colunas
```

```python
bad_years = movies["Ano"].value_counts().nsmallest(5) #RESPOSTA - separando cada ano que teve publicação de filmes e contando quantas publicações teve de filme nos respectivos anos
bad_years.head() #RESPOSTA.
```

**1.4. Quantos filmes que possuem avaliação maior ou igual a 4.7, considerando apenas os filmes avaliados na última data de avaliação do dataset?**

```python
large_rating = customer.query('Rating >= 4.7') #Separando apenas os filmes com notas maiores que 4.7 e atribuindo um nome a essa nova tabela
large_rating
```

```python
best_movies = large_rating.sort_values(['Date'], ascending=[False]) #Retornando os filmes organizados pela ordem de Data ('Date') da maior para a menor em uma nova tabela
best_movies
```

```python
best_movies.query('Date == "2005-12-31"').shape[0] #RESPOSTA - Como já descobrimos a maior data, agora retornamos quantos filmes foram publicados nessa data.RESPOSTA.
```
